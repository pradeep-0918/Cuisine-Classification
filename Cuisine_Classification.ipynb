***Task3: Cuisine Classification***

```python
import pandas as pd
import numpy as np
df = pd.read_csv("/content/Dataset .csv")
print("Dataset is loaded Successfully")
```

```python
print("Handling Missing Values")
df.isnull().sum()
```

```python
print("Fill missing Value using by taking mode()")
df['Cuisines'] = df['Cuisines'].fillna(df['Cuisines'].mode()[0])
df.isnull().sum()
```

```python
from sklearn.preprocessing import MultiLabelBinarizer, StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
print("Coverting the Cuisines -> Categorical data into numerical data by using MLb")
df['Cuisines_List'] = df['Cuisines'].apply(lambda x: x.split(', '))
mlb = MultiLabelBinarizer()
y = mlb.fit_transform(df['Cuisines_List'])
print(f"\nTarget variable (y)_shape): {y.shape}")
print(f"First 5 rows of encoded target (y):\n{y[:5]}")
print(f"Unique cuisine classes (first 10): {mlb.classes_[:10].tolist()}...")
```

```python
columns_to_drop_from_x = [
    'Restaurant ID', 'Restaurant Name', 'Address', 'Locality', 'Locality Verbose',
    'Cuisines', 'Cuisines_List', 'Switch to order menu', 'Currency',
    'Aggregate rating', 'Rating color', 'Rating text', # These are outcomes, not predictors of cuisine
]
x = df.drop(columns=columns_to_drop_from_x, errors='ignore')

print(f"\nFeatures (X) shape: {x.shape}")
print("First 5 rows of features (X):")
print(x.head())
```

```python
print("spliting up the dataset into two categories")
numerical_cols = ['Longitude', 'Latitude', 'Average Cost for two', 'Votes', 'Price range']
categorical_cols = ['Country Code', 'City', 'Has Table booking', 'Has Online delivery', 'Is delivering now']
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_cols),
        ('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=False), categorical_cols)
    ],
    remainder='passthrough'
)

print("\nColumnTransformer (preprocessor) defined for features (X).")
```

```python
from sklearn.multiclass import OneVsRestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import jaccard_score, hamming_loss, accuracy_score, precision_recall_fscore_support, classification_report
```

```python
logistic_reg_classifier = OneVsRestClassifier(LogisticRegression(solver='liblinear', max_iter=200, random_state=42))
logistic_reg_classifier.fit(x_train_transformed, y_train)

rf_classifier = OneVsRestClassifier(RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1))
rf_classifier.fit(x_train_transformed, y_train)
```

```python
y_pred_log_reg = logistic_reg_classifier.predict(x_test_transformed)
jaccard_log_reg = jaccard_score(y_test, y_pred_log_reg, average='samples')
hamming_loss_log_reg = hamming_loss(y_test, y_pred_log_reg)
accuracy_log_reg = accuracy_score(y_test, y_pred_log_reg)
precision_micro_log_reg, recall_micro_log_reg, f1_micro_log_reg, _ = precision_recall_fscore_support( y_test, y_pred_log_reg, average='micro', zero_division=0)
precision_macro_log_reg, recall_macro_log_reg, f1_macro_log_reg, _ = precision_recall_fscore_support(y_test, y_pred_log_reg, average='macro', zero_division=0)
```

```python
y_pred_rf = rf_classifier.predict(x_test_transformed)
jaccard_rf = jaccard_score(y_test, y_pred_rf, average='samples')
hamming_loss_rf = hamming_loss(y_test, y_pred_rf)
accuracy_rf = accuracy_score(y_test, y_pred_rf)
precision_micro_rf, recall_micro_rf, f1_micro_rf, _ = precision_recall_fscore_support( y_test, y_pred_rf, average='micro', zero_division=0)
precision_macro_rf, recall_macro_rf, f1_macro_rf, _ = precision_recall_fscore_support(y_test, y_pred_rf, average='macro', zero_division=0)
cuisine_labels = mlb.classes_
```

```python
log_reg_report = classification_report(y_test, y_pred_log_reg, target_names=cuisine_labels, zero_division=0)
rf_report = classification_report(y_test, y_pred_rf, target_names=cuisine_labels, zero_division=0)

precision_lr_per_label, recall_lr_per_label, f1_lr_per_label, support_lr_per_label = \
precision_recall_fscore_support(y_test, y_pred_log_reg, average=None, zero_division=0)

precision_rf_per_label, recall_rf_per_label, f1_rf_per_label, support_rf_per_label = \
precision_recall_fscore_support(y_test, y_pred_rf, average=None, zero_division=0)
```
